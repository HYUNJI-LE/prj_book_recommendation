```python
import time
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
from selenium import webdriver
```

```python
options = webdriver.ChromeOptions()
#options.add_argument('headless')
options.add_argument('disable-gpu')
options.add_argument('lang=ko_KR')
driver = webdriver.Chrome('./chromedriver', options=options)
```

```python
titles = []
contents = []
#100, 110, 160, 170, 120,/ 190,  180, 250, 210, 200, 310, 320, 130, 140,/ 150, 270, 230, 240, 280, 300, 330, 290


book_list =[150, 270, 230, 240, 280, 300, 330, 290]

for i in book_list:
    url = 'https://book.naver.com/category/index.nhn?cate_code={0}'.format(i)
    driver.get(url)
    time.sleep(0.5)
    review_len = int(driver.find_element_by_xpath('//*[@id="content"]/div[4]/ul/li[3]/em').text.replace(',', '')) #리뷰갯수
    print(review_len)
    for h in range(1,int(np.ceil(review_len/20)+1)):
        try:
            for k in range(0,20):

                try:
                    url = 'https://book.naver.com/category/index.nhn?cate_code={0}&tab=recommend&list_type=list&sort_type=salecount&page={1}'.format(i,h)
                    driver.get(url)
                    time.sleep(0.5)

                    title = driver.find_element_by_xpath('//*[@id="book_title_{}"]/a[1]'.format(k)).text #책 제목 가져오는 코드
                    print(title)

                    driver.find_element_by_xpath('//*[@id="book_title_{}"]/a[1]'.format(k)).click() #책 제목 클릭
                    time.sleep(0.5)

                    content = driver.find_element_by_xpath('//*[@id="content"]').text #서평
                    time.sleep(1.0)

                    titles.append(title)
                    contents.append(content)

                    driver.back() #이전 페이지 가기
                    time.sleep(0.5)

                except:
                    print('exception')
        except:
            break

driver.close()
```

```python
print(len(titles))
print(len(contents))
```

```python
df_review = pd.DataFrame({'title':titles, 'content':contents})
print(df_review.head())
print(df_review.tail())
print(df_review.info())
```

```python
df_review.to_csv('./crawling/daum_book2.csv')
```
