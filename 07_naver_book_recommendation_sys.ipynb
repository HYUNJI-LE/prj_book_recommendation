```python
import pandas as pd
pd.set_option('display.max_row',500)
pd.set_option('display.max_columns',100)
pd.set_option('display.unicode.east_asian_width',True) #한글
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
#문장간의 유사도를 구하는 방법 tfidf
#tf: 한문장에 몇번나왔는지(중요)
#idf: 모든 문장에 몇번나왔는지에 대한 역수(안중요)
df_review_one_sentence = pd.read_csv('./crawling/cleaned_daum_book.csv', index_col=0)
print(df_review_one_sentence.head(50))
print(df_review_one_sentence.info())
#영화 n개와 영화의 리뷰 n개
```

```python
#코사인 유사도로 가장 유사한 영화 n가지를 출력-> 두개의 벡터가 이루고 있는 값과 그 각도가 필요 ->1이면 유사, -1이면 반대, 0이면 상관이 없음 
def getReccomendation(cosine_sim): 
    simScores = list(enumerate(cosine_sim[-1]))
    #enumaerate : 인덱스를 붙여줌(어떤 영화인지 찾고 나중에 추천해주기 위해)
    #enumerate 객체로 나오기 때문에 리스트로 바꿔줌 (0을 줘도 상관 없음)
    simScores = sorted(simScores, key=lambda x: x[1], reverse=True) #내림차 순으로 정렬
    simScores = simScores[1:31] #30개 추천
    movieidx = [i[0] for i in simScores] #simScores에 큰 순서대로 코사인 세터값이 들어가게 되고 영화인덱스(i)에서 첫번째의 인덱스를 뽑아옴
    RecMovielist = df_review_one_sentence.iloc[movieidx] #뽑힌 인덱스로 영화제목과 리뷰를 찾아옴
    return RecMovielist.title #영화제목과 리뷰를 리턴
```

```python
movie_idx = df_review_one_sentence[df_review_one_sentence['title']=='오만과 편견'].index[0]
#영화 제목을 입력해서 비슷한 영화를 찾을 때 이름을 넣으면 인덱스가 나오도록(영화 인덱스를 찾아냄)
```

```python
#Tfidf 메트릭스로 벡터값으로 변환
Tfidf = TfidfVectorizer(sublinear_tf=True)
Tfidf_matrix = Tfidf.fit_transform(df_review_one_sentence['cleaned_content'])
print(Tfidf_matrix.shape)
print(Tfidf_matrix)
#한문장에서 특정 단어가 몇번나오는지, 전체의 리뷰에서 특정 단어가 몇번 나오는지에 대한 로그 값을 구함 
#588개의 로우(모든 리뷰들의 갯수),64332개의 컬럼(모든 형태소들의 갯수)
#첫번째 형태소의 Tfidf 값 
#0번 로우에 9484번 컬럼의 값이 0.0014 -> 0번 로우에 포함되어 있는 형태소들만 표현(0이 아닌 값이 있는 것들만 보여줌)
```

```python
cosine_sim = linear_kernel(Tfidf_matrix[movie_idx], Tfidf_matrix)
print(cosine_sim.shape) 
#영화 하나(movie_idx)에 대한 588개의 코사인세터값이 나옴 (두개 벡터 간의 각도)
#linear_kernel : 코사인 유사도를 구해줌 
```

```python
#특정 영화와 모든 영화들 간의 유사도 점수 
print(cosine_sim)
```
